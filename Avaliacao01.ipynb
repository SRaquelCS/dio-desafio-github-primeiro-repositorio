**(1,0 ponto)** 1.c) É importante acompanhar em tempo real a evolução dos retornos dos investimentos, principalmente no caso de ações.
*  **Retorno percentual** é a diferença entre o `valor atual` e o `valor pago`: $\frac{atual - pago}{pago} \times 100\%$
> Crie uma função da qual:
> *  Os argumentos de entrada sejam `pago` e `atual`
> * Retorne o **retorno percentual** do investimento

Teste a função sabendo que o cliente investiu 153,67 reais e no momento, o valor de mercado deste investimento é 167,28 reais.

```python
# Questao 1
# 1.c)

# Funcao
def calcular_retorno_percentual(pago, atual):
    try:
        retorno_percentual = ((atual - pago) / pago) * 100
        return retorno_percentual
    except ZeroDivisionError:
        print("Erro: O valor pago não pode ser zero.")

# Teste da função
valor_pago = 153.67
valor_atual = 167.28

retorno_percentual = calcular_retorno_percentual(valor_pago, valor_atual)
print(f"O retorno percentual do investimento é: {retorno_percentual:.2f}%")

# Teste
valor_pago = 153.67
valor_atual = 167.28


```

> _O Mercosul, ou Mercado Comum do Sul, é um bloco econômico e político sul-americano criado em 1991. Seu objetivo é criar um mercado comum entre os países membros, com a livre circulação de bens, serviços, pessoas, capitais e informações. O bloco também busca promover a cooperação em áreas como educação, cultura, ciência e tecnologia._
>
> _Saiba mais: https://www.mercosur.int/pt-br/quem-somos/paises-do-mercosul/_
>
> Os países participantes são:
> * Países parte
>   * Argentina (`AR`)
>   * Brasil (`BR`)
>   * Paraguai (`PY`)
>   * Uruguai (`UY`)
>   * Venezuela* (`VE`)
> * Países associados
>   * Bolívia (`BO`)
>   * Chile (`CL`)
>   * Colômbia (`CO`)
>   * Equador (`EC`)
>   * Guiana (`GY`)
>   * Peru (`PE`)
>   * Suriname (`SR`)
>
> _Um importante indicador para o planejamento econômico e social destes países é a **expectativa de vida ao nascer** (`SP.DYN.LE00.IN`). Um país com uma população com expectativa de vida alta terá uma população mais ativa e produtiva, o que pode contribuir para o crescimento econômico e atrair melhores investimentos externos._

**(1,0 ponto)** 4.a) Utilizando o módulo `wb` do pacote `pandas_datareader`, obtenha a _expectativa de vida ao nascer_ (`SP.DYN.LE00.IN`) de todos os países do Mercosul para o ano de 2021.

*   Reinicie os índices das linhas
*   Atribua a variável `country` como o novo índice das linhas
*   Utilizando o método `.loc`, selecione somente a coluna `SP.DYN.LE00.IN`

```python
# Questao 4
# 4.a)
import pandas_datareader as pdr

# Lista de países do Mercosul
paises_mercosul = ['AR', 'BR', 'PY', 'UY', 'VE', 'BO', 'CL', 'CO', 'EC', 'GY', 'PE', 'SR']

# Obtendo a expectativa de vida ao nascer para o ano de 2021
dados_expectativa_vida = pdr.wb.download(indicator='SP.DYN.LE00.IN', country=paises_mercosul, start=2021, end=2021)

# Reiniciando os índices das linhas
dados_expectativa_vida.reset_index(inplace=True)

# Atribuindo 'country' como o novo índice das linhas
dados_expectativa_vida.set_index('country', inplace=True)

# Utilizando o método .loc para selecionar somente a coluna SP.DYN.LE00.IN
expectativa_vida_2021 = dados_expectativa_vida.loc[:, 'SP.DYN.LE00.IN']

print(expectativa_vida_2021)
```

**(1,0 ponto)** 4.b) Construa um gráfico de barras do indicador em questão para estes países.

*   Adicione adequadamente título e rótulos nos eixos

```python
# Questao 4
# 4.b)
import pandas_datareader as pdr
import matplotlib.pyplot as plt

# Lista de países do Mercosul
paises_mercosul = ['AR', 'BR', 'PY', 'UY', 'VE']

# Obtendo a expectativa de vida ao nascer para o ano de 2021
dados_expectativa_vida = pdr.data.WorldBankReader('SP.DYN.LE00.IN', start=2021, end=2021, country=paises_mercosul).read()

# Reiniciando os índices das linhas
dados_expectativa_vida.reset_index(inplace=True)

# Atribuindo 'country' como o novo índice das linhas
dados_expectativa_vida.set_index('country', inplace=True)

# Selecionando somente a coluna SP.DYN.LE00.IN
expectativa_vida_2021 = dados_expectativa_vida['SP.DYN.LE00.IN']

# Construindo o gráfico de barras
plt.figure(figsize=(10, 6))
expectativa_vida_2021.plot(kind='bar', color='skyblue')
plt.title('Expectativa de Vida ao Nascer (2021) - Países do Mercosul')
plt.xlabel('Países do Mercosul')
plt.ylabel('Expectativa de Vida ao Nascer')
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x para melhor legibilidade

# Exibindo o gráfico
plt.show()






```
